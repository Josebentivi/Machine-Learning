# Pergunta 1

```python
dados_filtrados1 = dados_brutos[['GROUP','CC_V_0','CC_V_5','IMC_V_0','IMC_V_5','PERCENT_MG_V_0','PERCENT_MG_V_5',
                                 'PERCENT_MM_V_0','PERCENT_MM_V_5','PESO_V_0','PESO_V_5','REACTANCIA_V_0',
                                 'REACTANCIA_V_5','RESISTENCIA_V_0','RESISTENCIA_V_5']]
dados_filtrados1.head(5)
```

```python
dados_filtrados1 = dados_filtrados1.dropna(axis=0)
dados_filtrados1.head(5)
```

```python
import scipy.stats as stats

def AdicionarValor(coluna1,coluna2, pretabela_resultados): 
    
    
    mediaInicial = coluna1.mean()
    desvioPadraoInicial = coluna1.std()
    v0 = str(round(mediaInicial,2))+" ("+str(round(desvioPadraoInicial,2))+")"
    mediaFinal = coluna2.mean()
    desvioPadraoFinal = coluna2.std()
    v5 = str(round(mediaFinal,2))+" ("+str(round(desvioPadraoFinal,2))+")"
    
    delta_media= mediaFinal-mediaInicial
    t,p = stats.ttest_rel(a=coluna1,b=coluna2)
    p = round(p,3)
    
    
    
    pretabela_resultados = pretabela_resultados.append({'V0': v0, 'V5': v5, 'p': p,'Delta M√©dia':delta_media}, ignore_index=True)
    return pretabela_resultados
```

```python
def GerandoResultadosPergunta1(dados_pergunta):
    tabela_resultados_pergunta1 = pd.DataFrame(columns=['V0','V5','p','Delta M√©dia'])
    
    
    tabela_resultados_pergunta1 = AdicionarValor(dados_pergunta['CC_V_0'],
                                             dados_pergunta['CC_V_5'],
                                            tabela_resultados_pergunta1)
    tabela_resultados_pergunta1 = AdicionarValor(dados_pergunta['IMC_V_0'],
                                             dados_pergunta['IMC_V_5'],
                                            tabela_resultados_pergunta1)
    tabela_resultados_pergunta1 = AdicionarValor(dados_pergunta['PERCENT_MG_V_0'],
                                             dados_pergunta['PERCENT_MG_V_5'],
                                            tabela_resultados_pergunta1)
    tabela_resultados_pergunta1 = AdicionarValor(dados_pergunta['PERCENT_MM_V_0'],
                                             dados_pergunta['PERCENT_MM_V_5'],
                                             tabela_resultados_pergunta1)
    tabela_resultados_pergunta1 = AdicionarValor(dados_pergunta['PESO_V_0'],
                                             dados_pergunta['PESO_V_5'],
                                             tabela_resultados_pergunta1)
    tabela_resultados_pergunta1 = AdicionarValor(dados_pergunta['REACTANCIA_V_0'],
                                             dados_pergunta['REACTANCIA_V_5'],
                                             tabela_resultados_pergunta1)
    tabela_resultados_pergunta1 = AdicionarValor(dados_pergunta['RESISTENCIA_V_0'],
                                             dados_pergunta['RESISTENCIA_V_5'],
                                             tabela_resultados_pergunta1)
    
    return tabela_resultados_pergunta1
```

```python
def TabelaCompleta1():
    grupo1_pergunta1 = dados_filtrados1.query('GROUP == 1')
    g1 = GerandoResultadosPergunta1(grupo1_pergunta1)
     
    grupo2_pergunta1 = dados_filtrados1.query('GROUP == 2')
    g2 = GerandoResultadosPergunta1(grupo2_pergunta1)
    
    grupo3_pergunta1 = dados_filtrados1.query('GROUP == 3')
    g3 = GerandoResultadosPergunta1(grupo3_pergunta1)
    
    tabelaNova = pd.DataFrame(columns=['p','F'])
    F, p = stats.f_oneway(grupo1_pergunta1['CC_V_0']-grupo1_pergunta1['CC_V_5'],
                          grupo2_pergunta1['CC_V_0']-grupo2_pergunta1['CC_V_5'],
                          grupo3_pergunta1['CC_V_0']-grupo3_pergunta1['CC_V_5']) 
    tabelaNova = tabelaNova.append({'p': round(p,3), 'F': F}, ignore_index=True)
    
    F, p = stats.f_oneway(grupo1_pergunta1['IMC_V_0']-grupo1_pergunta1['IMC_V_5'],
                          grupo2_pergunta1['IMC_V_0']-grupo2_pergunta1['IMC_V_5'],
                          grupo3_pergunta1['IMC_V_0']-grupo3_pergunta1['IMC_V_5']) 
    tabelaNova = tabelaNova.append({'p': round(p,3), 'F': F}, ignore_index=True)
    
    F, p = stats.f_oneway(grupo1_pergunta1['PERCENT_MG_V_0']-grupo1_pergunta1['PERCENT_MG_V_5'],
                          grupo2_pergunta1['PERCENT_MG_V_0']-grupo2_pergunta1['PERCENT_MG_V_5'],
                          grupo3_pergunta1['PERCENT_MG_V_0']-grupo3_pergunta1['PERCENT_MG_V_5']) 
    tabelaNova = tabelaNova.append({'p': round(p,3), 'F': F}, ignore_index=True)
    
    F, p = stats.f_oneway(grupo1_pergunta1['PERCENT_MM_V_0']-grupo1_pergunta1['PERCENT_MM_V_5'],
                          grupo2_pergunta1['PERCENT_MM_V_0']-grupo2_pergunta1['PERCENT_MM_V_5'],
                          grupo3_pergunta1['PERCENT_MM_V_0']-grupo3_pergunta1['PERCENT_MM_V_5']) 
    tabelaNova = tabelaNova.append({'p': round(p,3), 'F': F}, ignore_index=True)
    
    F, p = stats.f_oneway(grupo1_pergunta1['PESO_V_0']-grupo1_pergunta1['PESO_V_5'],
                          grupo2_pergunta1['PESO_V_0']-grupo2_pergunta1['PESO_V_5'],
                          grupo3_pergunta1['PESO_V_0']-grupo3_pergunta1['PESO_V_5']) 
    tabelaNova = tabelaNova.append({'p': round(p,3), 'F': F}, ignore_index=True)
    
    F, p = stats.f_oneway(grupo1_pergunta1['REACTANCIA_V_0']-grupo1_pergunta1['REACTANCIA_V_5'],
                          grupo2_pergunta1['REACTANCIA_V_0']-grupo2_pergunta1['REACTANCIA_V_5'],
                          grupo3_pergunta1['REACTANCIA_V_0']-grupo3_pergunta1['REACTANCIA_V_5']) 
    tabelaNova = tabelaNova.append({'p': round(p,3), 'F': F}, ignore_index=True)
    
    F, p = stats.f_oneway(grupo1_pergunta1['RESISTENCIA_V_0']-grupo1_pergunta1['RESISTENCIA_V_5'],
                          grupo2_pergunta1['RESISTENCIA_V_0']-grupo2_pergunta1['RESISTENCIA_V_5'],
                          grupo3_pergunta1['RESISTENCIA_V_0']-grupo3_pergunta1['RESISTENCIA_V_5']) 
    tabelaNova = tabelaNova.append({'p': round(p,3), 'F': F}, ignore_index=True)
    


    g1 = g1.rename({"V0":"V0 Grupo1","V5":"V5 Grupo1","p":"p Grupo1","Delta M√©dia":"Delta M√©dia G1"},axis='columns')
    g2 = g2.rename({"V0":"V0 Grupo2","V5":"V5 Grupo2","p":"p Grupo2","Delta M√©dia":"Delta M√©dia G2"},axis='columns')
    g3 = g3.rename({"V0":"V0 Grupo3","V5":"V5 Grupo3","p":"p Grupo3","Delta M√©dia":"Delta M√©dia G3"},axis='columns')
    
    
    resultado = pd.concat([g1,g2,g3], axis=1, sort=False)
    
    resultado = resultado.rename({0: "CC", 1:"IMC",2:"PERCENT MG",3:"PERCENT MM",
                      4:"PESO",5:"REACTANCIA",6:"RESISTENCIA"}, axis='index')
    
    tabelaNova = tabelaNova.rename({0: "CC", 1:"IMC",2:"PERCENT MG",3:"PERCENT MM",
                      4:"PESO",5:"REACTANCIA",6:"RESISTENCIA"}, axis='index')
    
    return resultado,tabelaNova
```

```python
resultado_tStudentPareado,tabelaNova= TabelaCompleta1()
```

```python
def GerandoResultadosPergunta4(dados_pergunta):
    tabela_resultados_pergunta4 = pd.DataFrame(columns=['V0','V5','p','Delta M√©dia'])
    
    tabela_resultados_pergunta4 = AdicionarValor(dados_pergunta['ALANINA_AMINOTRANSFERASE_V_0'],
                                             dados_pergunta['ALANINA_AMINOTRANSFERASE_V_5'],
                                            tabela_resultados_pergunta4)
    tabela_resultados_pergunta4 = AdicionarValor(dados_pergunta['ASPARTATO_AMINO_TRANSFERASE_V_0'],
                                             dados_pergunta['ASPARTATO_AMINO_TRANSFERASE_V_5'],
                                            tabela_resultados_pergunta4)
    tabela_resultados_pergunta4 = AdicionarValor(dados_pergunta['FOSFATASE_ALCALINA_V_0'],
                                             dados_pergunta['FOSFATASE_ALCALINA_V_5'],
                                            tabela_resultados_pergunta4)
    tabela_resultados_pergunta4 = AdicionarValor(dados_pergunta['GAMA_GLUTAMIL_TRANSFERASE_V_0'],
                                             dados_pergunta['GAMA_GLUTAMIL_TRANSFERASE_V_5'],
                                            tabela_resultados_pergunta4)
    tabela_resultados_pergunta4 = AdicionarValor(dados_pergunta['CC_V_0'],
                                             dados_pergunta['CC_V_5'],
                                            tabela_resultados_pergunta4)
    tabela_resultados_pergunta4 = AdicionarValor(dados_pergunta['GLICOSE_V_0'],
                                             dados_pergunta['GLICOSE_V_5'],
                                            tabela_resultados_pergunta4)
    tabela_resultados_pergunta4 = AdicionarValor(dados_pergunta['INSULINA_V_0'],
                                             dados_pergunta['INSULINA_V_5'],
                                            tabela_resultados_pergunta4)
    

    return tabela_resultados_pergunta4
```

```python
def TabelaCompleta4():
    grupo1_pergunta4 = dados_filtrados4.query('GROUP == 1')
    g1 = GerandoResultadosPergunta4(grupo1_pergunta4)
     
    grupo2_pergunta4 = dados_filtrados4.query('GROUP == 2')
    g2 = GerandoResultadosPergunta4(grupo2_pergunta4)
    
    grupo3_pergunta4 = dados_filtrados4.query('GROUP == 3')
    g3 = GerandoResultadosPergunta4(grupo3_pergunta4)
    
    
    tabelaNova = pd.DataFrame(columns=['p','F'])
    F, p = stats.f_oneway(grupo1_pergunta4['ALANINA_AMINOTRANSFERASE_V_0']-grupo1_pergunta4['ALANINA_AMINOTRANSFERASE_V_5'],
                          grupo2_pergunta4['ALANINA_AMINOTRANSFERASE_V_0']-grupo2_pergunta4['ALANINA_AMINOTRANSFERASE_V_5'],
                          grupo3_pergunta4['ALANINA_AMINOTRANSFERASE_V_0']-grupo3_pergunta4['ALANINA_AMINOTRANSFERASE_V_5']) 
    tabelaNova = tabelaNova.append({'p': round(p,3), 'F': F}, ignore_index=True)
    
    F, p = stats.f_oneway(grupo1_pergunta4['ASPARTATO_AMINO_TRANSFERASE_V_0']-grupo1_pergunta4['ASPARTATO_AMINO_TRANSFERASE_V_5'],
                          grupo2_pergunta4['ASPARTATO_AMINO_TRANSFERASE_V_0']-grupo2_pergunta4['ASPARTATO_AMINO_TRANSFERASE_V_5'],
                          grupo3_pergunta4['ASPARTATO_AMINO_TRANSFERASE_V_0']-grupo3_pergunta4['ASPARTATO_AMINO_TRANSFERASE_V_5']) 
    tabelaNova = tabelaNova.append({'p': round(p,3), 'F': F}, ignore_index=True)
    
    F, p = stats.f_oneway(grupo1_pergunta4['FOSFATASE_ALCALINA_V_0']-grupo1_pergunta4['FOSFATASE_ALCALINA_V_5'],
                          grupo2_pergunta4['FOSFATASE_ALCALINA_V_0']-grupo2_pergunta4['FOSFATASE_ALCALINA_V_5'],
                          grupo3_pergunta4['FOSFATASE_ALCALINA_V_0']-grupo3_pergunta4['FOSFATASE_ALCALINA_V_5']) 
    tabelaNova = tabelaNova.append({'p': round(p,3), 'F': F}, ignore_index=True)
    
    F, p = stats.f_oneway(grupo1_pergunta4['GAMA_GLUTAMIL_TRANSFERASE_V_0']-grupo1_pergunta4['GAMA_GLUTAMIL_TRANSFERASE_V_5'],
                          grupo2_pergunta4['GAMA_GLUTAMIL_TRANSFERASE_V_0']-grupo2_pergunta4['GAMA_GLUTAMIL_TRANSFERASE_V_5'],
                          grupo3_pergunta4['GAMA_GLUTAMIL_TRANSFERASE_V_0']-grupo3_pergunta4['GAMA_GLUTAMIL_TRANSFERASE_V_5']) 
    tabelaNova = tabelaNova.append({'p': round(p,3), 'F': F}, ignore_index=True)
    
    F, p = stats.f_oneway(grupo1_pergunta4['CC_V_0']-grupo1_pergunta4['CC_V_5'],
                          grupo2_pergunta4['CC_V_0']-grupo2_pergunta4['CC_V_5'],
                          grupo3_pergunta4['CC_V_0']-grupo3_pergunta4['CC_V_5']) 
    tabelaNova = tabelaNova.append({'p': round(p,3), 'F': F}, ignore_index=True)
    
    F, p = stats.f_oneway(grupo1_pergunta4['GLICOSE_V_0']-grupo1_pergunta4['GLICOSE_V_5'],
                          grupo2_pergunta4['GLICOSE_V_0']-grupo2_pergunta4['GLICOSE_V_5'],
                          grupo3_pergunta4['GLICOSE_V_0']-grupo3_pergunta4['GLICOSE_V_5']) 
    tabelaNova = tabelaNova.append({'p': round(p,3), 'F': F}, ignore_index=True)
    
    F, p = stats.f_oneway(grupo1_pergunta4['INSULINA_V_0']-grupo1_pergunta4['INSULINA_V_5'],
                          grupo2_pergunta4['INSULINA_V_0']-grupo2_pergunta4['INSULINA_V_5'],
                          grupo3_pergunta4['INSULINA_V_0']-grupo3_pergunta4['INSULINA_V_5']) 
    tabelaNova = tabelaNova.append({'p': round(p,3), 'F': F}, ignore_index=True)
    

    g1 = g1.rename({"V0":"V0 Grupo1","V5":"V5 Grupo1","p":"p Grupo1","Delta M√©dia":"Delta M√©dia G1"},axis='columns')
    g2 = g2.rename({"V0":"V0 Grupo2","V5":"V5 Grupo2","p":"p Grupo2","Delta M√©dia":"Delta M√©dia G2"},axis='columns')
    g3 = g3.rename({"V0":"V0 Grupo3","V5":"V5 Grupo3","p":"p Grupo3","Delta M√©dia":"Delta M√©dia G3"},axis='columns')
    
    
    resultado = pd.concat([g1,g2,g3], axis=1, sort=False)
    
    resultado = resultado.rename({0: "ALANINA AMINOTRANSFERASE", 1:"ASPARTATO AMINO TRANSFERASE",
                                  2:"FOSFATASE ALCALINA",3:"GAMA GLUTAMIL TRANSFERASE",4:"CC",
                                  5:"GLICOSE",6:"INSULINA"}, axis='index')
    
    tabelaNova = tabelaNova.rename({0: "ALANINA AMINOTRANSFERASE", 1:"ASPARTATO AMINO TRANSFERASE",
                                  2:"FOSFATASE ALCALINA",3:"GAMA GLUTAMIL TRANSFERASE",4:"CC",
                                  5:"GLICOSE",6:"INSULINA"}, axis='index')
    
    return resultado,tabelaNova
```
